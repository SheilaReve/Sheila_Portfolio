**do analysis on all of the nutrient data.**

-assemble the nutrients for each food into a single large table

-1) convert each list of food nutrients to a DataFrame

-2) add a column for the food id

-3) append the dataframe to a list

-4) concatenate all the lists together (with concat)

```python
nutrients=[]
```

```python
for rec in db:
    fnuts=pd.DataFrame(rec['nutrients'])
    fnuts['id']=rec['id']
    nutrients.append(fnuts)
    
```

```python
nutrients=pd.concat(nutrients,ignore_index=True)
```
